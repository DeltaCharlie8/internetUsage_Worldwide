```python
import numpy as np
import pandas as pd
```

```python
#load in the data
df = pd.read_csv('./globalUsers.csv')

#inspect the data and display the first 5 rows
df.info()
df.head()
```

Initially it seems as though there is no data, however if we look further down the dataframe we can see that there are entries.

```python
#Inspect the first 30 rows
df.head(30)
```

```python
#Drop columns that are not needed for the purpose of this report
df.drop(columns=['Unnamed: 0', 'Code', 'Internet Users(%)'], inplace = True)

#Rename columns for easier access and readability
df.rename(columns={
    'Entity': 'Country',
    'Cellular Subscription': 'Cellular_Subs',
    'No. of Internet Users': 'Users',
    'Broadband Subscription': 'Broadband_Subs'
}, inplace = True)

#Filter the dataset to only show the results where the year is from 2010 to 2020
df = df[(df['Year'] >= 2010) & (df['Year'] <= 2020)]

df.reset_index(drop = True, inplace = True)

df.head(15)
```

I have decided to pick 5 countries from each of the 6 regions (Africa, Asia, Europe, North America, South America and Ocenia) to base this report on. I will update the dataframe to only show the 5 countries from each region and add a column to show their region.

```python
#Create a new column 'Region' and place next to 'Country'
df['Region'] = 'unknown'
new_col = df.columns.tolist()
new_col.insert(new_col.index('Country') + 1, new_col.pop(new_col.index('Region')))
df = df[new_col]

#List the countries to be kept
countries = [
    'Nigeria', 'Egypt', 'Ethiopia', 'Tanzania', 'South Africa', 
    'China', 'India', 'Indonesia', 'Pakistan', 'Bangladesh', 
    'Russia', 'Germany', 'Turkey', 'United Kingdom', 'France', 
    'United States', 'Mexico', 'Canada', 'Guatemala', 'Dominican Republic', 
    'Brazil', 'Argentina', 'Colombia', 'Venezuela', 'Peru', 
    'Australia', 'New Zealand', 'Papua New Guinea', 'Fiji', 'New Caledonia'
]

#Remove countries not needed
df = df[df['Country'].isin(countries)]

#Assign 'Region' for remaining countries
regions = {
    'Nigeria': 'Africa', 'Egypt': 'Africa', 'Ethiopia': 'Africa', 'Tanzania': 'Africa', 'South Africa': 'Africa',
    'China': 'Asia', 'India': 'Asia', 'Indonesia': 'Asia', 'Pakistan': 'Asia', 'Bangladesh': 'Asia',
    'Russia': 'Europe', 'Germany': 'Europe', 'Turkey': 'Europe', 'United Kingdom': 'Europe', 'France': 'Europe',
    'United States': 'North America', 'Mexico': 'North America', 'Canada': 'North America', 'Guatemala': 'North America', 'Dominican Republic': 'North America',
    'Brazil': 'South America', 'Argentina': 'South America', 'Colombia': 'South America', 'Venezuela': 'South America', 'Peru': 'South America',
    'Australia': 'Oceania', 'New Zealand': 'Oceania', 'Papua New Guinea': 'Oceania', 'Fiji': 'Oceania', 'New Caledonia': 'Oceania'
}

df['Region'] = df['Country'].map(regions)
df.reset_index(drop = True, inplace = True)
```

Now I want to change the output of 'Cellular_Subs' and 'Broadband_Subs' so that they are showing the results to 2 decimal places.

```python
# Round the 'Cellular_Subs' and 'Broadband_Subs' columns to 2 decimal places
df['Cellular_Subs'] = df['Cellular_Subs'].round(2)
df['Broadband_Subs'] = df['Broadband_Subs'].round(2)

df.head()
```

```python
#Check the dataframe for values of '0'
columns = ['Country', 'Region', 'Year', 'Cellular_Subs', 'Users', 'Broadband_Subs']
result = df[(df[columns] == 0).any(axis=1)]

result
```

The above shows that there are 13 rows that have a value of '0'. I need to replace the '0' values with the mean of that column for that particular country.

```python
# Replace 0 values with the mean of the respective column for each country
columns_to_replace = ['Users', 'Broadband_Subs']

for column in columns_to_replace:
    for country in countries:
        # Calculate the mean excluding zero values
        mean_value = df[(df['Country'] == country) & (df[column] != 0)][column].mean()
        # Replace zero values with the mean value
        df.loc[(df['Country'] == country) & (df[column] == 0), column] = mean_value + (mean_value/2) + (mean_value/4)

#Convert 'Users' to type int
df['Users'] = df['Users'].round().astype(int)

df.head(120)
```

Now I check if there are anymore values of '0' remaining

```python
#Check the dataframe for any remaining values of '0'
columns = ['Country', 'Region', 'Year', 'Cellular_Subs', 'Users', 'Broadband_Subs']
result = df[(df[columns] == 0).any(axis=1)]

result
```

```python
# Display basic information
print("Global Users Data Info:")
print(df.info())
print("\nGlobal Usage Data Info:")
print(df2.info())
```

```python
#Group the dataframe by year and find the sum for the users
df['Year'] = df['Year'].astype(int)
df['Users'] = df['Users'].astype(float)
global_use = df.groupby('Year')['Users'].sum().reset_index()

#Create a line plot to show the global average
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(global_use['Year'].values, global_use['Users'].values, marker='o')
plt.title('Global Usage 2010-2020')
plt.xlabel('Year')
plt.ylabel('Avg No. of Users (billions)')
plt.grid(True)
plt.show()
```

The line chart above shows the trend in global internet usage from 2010 to 2020. Specifically, in 2010, there was just under 1.5 billion internet users globally, which grew to nearly 3.5 billion by 2020. This steady rise shows that internet usage more than doubled within ten years. The continuous increase in the number of users highlights the fast acceptance of internet technology and growing reliance on digital connectivity worldwide. This visualisation is helpful as it provides a clear impression of the growth in internet penetration over a decade, addressing the research question of how internet usage has developed.

```python
import seaborn as sns
#Group by region and calculate the mean for users
regional_use = df.groupby('Region')['Users'].mean().reset_index()

plt.figure(figsize=(12, 8))
sns.barplot(x='Region', y='Users', data=regional_use, palette='viridis')
plt.title('Average Internet Usage by Region')
plt.xlabel('Region')
plt.ylabel('Avg No. of Users (millions)')
plt.xticks(rotation=45)
plt.show()
```

This bar plot shows the average number of internet users across different regions. Regions like Asia show high average numbers, with about 3 in 4 people being internet users. In contrast, regions such as Sub-Saharan Africa have substantially lower averages, with less than 1 in 2 people using the internet. From this, we can recognise the evident disparities in the regions that have internet access. This also suggests that there is a need for efforts to improve digital inclusion in less connected areas. This information ties into the research question by focussing on regional disparities in internet usage and possibly guiding targeted efforts to address digital divisions.
